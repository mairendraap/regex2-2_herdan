Variabel yang digunakan adalah variabel "isi" karena variabel tersebut digunakan untuk menyimpan hasil file.read dari file berita.txt, yang nantinya akan dipanggil kembali dengan print.

```python
import re

with open("datasets/berita.txt", "r") as f:
    isi = f.read()

pattern = re.compile(r"Prof\..+?([A-Z][a-z]+(?:\s[A-Z][a-z]+)+)")
matches = re.findall(pattern, isi)

sorted_matches = sorted(set(matches))
print(sorted_matches)
```

Sebenarnya ini sama saja seperti metode regex yang telah dilakukan di kelas praktikum, yang membedakan disini hanyalah pattern yang lebi rumit karena diminta untuk mencari atau menampilkan nama dosen tanpa gelar, dimana letak kesulitannya disini adalah nama dari dosen atau profesor disini diawali dengan gelar dan diakhiri dengan gelar juga, dan gelarnya terdapat tanda koma atau titik. Disitu saya menggunakan pattern kompleks yang mencari Prof dulu untuk di depan dan \. setelahnya juga iku, setelahnya saya grouping apa saja yang mau dicari setelah dari pattern tersebut seperti A-Z, a-z, :, dan . juga. Memang lumayan kompleks dan susah untuk dijelaskan namun karena grouping sebagai penghilang atau menghilangkan tanda special dalam regex. Sama seperti sebelumnya saya akan melihat terlebih dahulu apa isi dari file yang akan kita eksekusi, disini saya sengaja menggunakan pandas saja agar lebih simple daripada csv, karna ada kelebihan dan kekurangannya masing-masing. Saya import library dan menggunakan variabel filepath untuk menyimpan path file reviews_dataset_updated.csv, agar saat saya panggil di fungsi pd.read_csv saya tinggal panggil variabelnya. Dan setelahnya saya print.

```python
import csv
import re

with open('datasets/reviews_dataset_updated.csv', 'r', encoding='utf-8') as infile, \
     open('cleaned.csv', 'w', encoding='utf-8', newline='') as outfile:
    
    reader = csv.DictReader(infile)
    writer = csv.DictWriter(outfile, fieldnames=['review_text', 'punctuation_and_digit_removal', 'casefolding', 'sentiment'])
    writer.writeheader()
    
    for row in reader:
        cleaned = re.sub(r'[^a-zA-Z\s]', '', row['review_text'])
        cleaned = re.sub(r'\s+', ' ', cleaned).strip()
        
        row['punctuation_and_digit_removal'] = cleaned
        row['casefolding'] = cleaned.lower()
        
        writer.writerow(row)
```

Nah, ini lumayan tricky. Karena sebelumnya rata2 kita menggunakan library pandas untuk akses file csv, disini saya langsung menggunakan library yg sudah ada yaitu import csv untuk melakukan eksekusi kode, dengan ini kita dapat langsung melakukan rekayasa isi file dan membuat file baru misal cleaned.csv untuk hasil akhir atau output yang sudah kita eksekusi, disana terdapat 2 alias yaitu infile dan outfile, sama seperti maksudnya infile adalah file masuk yang kita read dan outfile adalah output, jadi kita masukkan saja di variabel reader dan write masing2. dan gunakan syntax dari libary csv, seperti csv.DictReader atau Dictionary Reader dan Dictionary Write, karena kita akan rewrite sesuai instruksi yaitu, punctuation digit removal, casefolding dan sentiment. setelah itu kita sub atau replace jika pada string biasa. pattern cleaned = re.sub(r'[^a-zA-Z\s]', '', row['review_text']) yang digunakan yakni mencari karakter yang bukan merupakan huruf dan spasi dan menghapus semua tanda baca, dan yang ini cleaned = re.sub(r'\s+', ' ', cleaned).strip() untuk merapikan spasi saja sih sebenernya. dan untuk bawahnya itu untuk menambahkan di row atau baris isi file yang akan menjadi output yakni cleaned.csv gampangnya seperti append atau menyisipkan isi baru, dan tinggal write aja per row.
